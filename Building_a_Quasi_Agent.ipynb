````python
import re
````

````python
def extract_code(text: str) -> str:
    """Извлекает код из блоков ```python ... ```"""
    match = re.search(r"```python\s*(.*?)\s*```", text, re.DOTALL)
    return match.group(1) if match else text
````

````python
from litellm import completion
from typing import List, Dict


def generate_response(messages: List[Dict]) -> str:
    """Call LLM to get response"""
    response = completion(
        model="gemini/gemini-2.5-flash",
        messages=messages,
        max_tokens=4096
    )
    return response.choices[0].message.content
````

````python
what_to_help_with = input("What function do you want to create?")

messages = [
    {"role": "system", "content": "You are a professional Python dev. Write ONLY the basic function code. Use functional programming if possible."},
    {"role": "user", "content": what_to_help_with}
]

response_1 = generate_response(messages)
print()
print(response_1)
````

````python
code_block_1 = extract_code(response_1)
print(code_block_1)
````

````python
messages.append({
    "role": "assistant",
    "content": code_block_1
})
````

````python
documentation_prompt = f"""
Add comprehensive documentation including:
Function description
Parameter descriptions
Return value description
Example usage
Edge cases
"""
messages.append(
    {"role": "user", "content": documentation_prompt}
)
````

````python
response_2 = generate_response(messages)
print(response_2)
````

````python
code_block_2 = extract_code(response_2)
print(code_block_2)
````

````python
messages.append({
    "role": "assistant",
    "content": code_block_2
})
````

````python
test_prompt = f"""
Add test cases using Python’s unittest framework
Tests should cover:
Basic functionality
Edge cases
Error cases
Various input scenarios
"""
messages.append({
    "role": "user",
    "content": test_prompt
})
````

````python
test_cases = generate_response(messages)
print(test_cases)
````

````python
code_block_3 = extract_code(test_cases)
print(code_block_3)
````
